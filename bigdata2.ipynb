```python
import requests
import json
import string
from bs4 import BeautifulSoup as bs
details =list()
tr=str.maketrans("","",string.punctuation)
url="https://www.daraz.com.np/"
response=requests.get(url)
home_page=bs(response.content,'lxml')
category_list=home_page.find_all('li',class_="menu-item")
for category in category_list:
    
    category_name=category.find('span',class_="nav-subTxt").get_text()
    
    clink=category.find('a').get('href')
    for i in range(1,26):
        
        response=requests.get(clink+str(i))
        cpage=bs(response.content,'lxml')
        #if(cpage.find('span',class_="total-products").get_text() == " 0 products found"):
            
            #print("found it")
            #break
        try:
            print(cpage.find('span',class_="total-products").get_text())
        except:
            print("0 result")
        phones=cpage.find_all('div', class_="sku -gallery ")
        for phone in phones:
            te=dict()
            name=phone.find('span', class_="name").get_text()
            te["name"]=name
            brand=phone.find('span',class_="brand").get_text()
            te["brand"]=brand
            price=phone.find('span',class_="price").get_text()
            price=price.translate(tr)
            price=price.split(" ")[1]
            price=price[1:-2]
            te["price"]=price
            try:
                rating=phone.find('div',class_="total-ratings").get_text()
                rating=rating.translate(tr)
                te["rating"]=rating
            except:
                te["rating"]=""
            link=phone.find('a').get('href')
            try:
                response=requests.get(link)
                print(link)
                next_page=bs(response.content,'lxml')
                specification=next_page.find('div',class_="osh-table -no-border").get_text()
                te["specification"]=specification
            except:
                te["specification"]="no data"
            details.append(te)
print(details)
file=open("bigdata2","w")
file.write(json.dumps(details))
    
```
